2\. _Social networking_ Please look at the graph below of social relationships. You have a mango farm, you need to find a mango buyer and sell him your mango that you have grown (just a small farm). To this end, our principle is to use our social network of friends separated by one degree, and then second degree friends (friends of friends), third degree friends (friends of friends of friends).

Use the depth first algorithm to search your network until you find a mango seller. 

We make an assumption that if the person's name ends with m, then it means the person must be the mango seller!


<img src="img/p3.png" width="500" height="400" >

```python
## write your answer here!
from collections import deque



graph = {}
graph["you"] = ["alice", "bob", "claire"]
graph["bob"] = ["anuj", "peggy"]
graph["alice"] = ["peggy"]
graph["claire"] = ["thom", "jonny"]
graph["anuj"] = []
graph["peggy"] = []
graph["thom"] = []
graph["jonny"] = []
def person_is_seller(name):
      return name[-1] == 'm'
    
def search(name):
    search_queue = deque() # creat queue
    search_queue += graph[name] # add all the neighbprs to the queue
    searched = [] # use this array to record  who are already search 
    while search_queue: # if queue is not empty
        person = search_queue.popleft() # choose one
        if not person in searched: # check if the person haven't been checked
            if person_is_seller(person): # juge if the person is the seller
                print (person + " is a mango seller!") 
                return True
            else:
                search_queue += graph[person] # if the person isn't a seller, add his friends to the queue
                searched.append(person) # marked this person
    return False # there isn't exist a mango seller

search("you")
```

Update your code to search for a particular filename.

What protocol did you use to select the next node to expand on the fringe?

```python
## write your answer here!
import os
 
def DFS_Dir(path):
    stack = []
    stack.append(path);
    while len(stack) > 0:
        tmp = stack.pop()# push out the current item and use LIFO
        FileList = os.listdir(tmp)
        for item in FileList:
            file_path= os.path.join(tmp, item)
            if(os.path.isdir(file_path)):# this item is dir or not
                stack.append(file_path)
            else:# juge if it is target file 
                if(str(file_path).endswith(".dll")):
                    print(item) 


try:
    DFS_Dir(r'C:\\')
except WindowsError:
        pass


```

Breadth First traverse：Use breadth traversal to traverse all the DLL files on the C disk and output the filename.

Update your code to search for a particular filename.

```python
## write your answer here!
import os
def BFS_Dir(path):
    queue = []
    queue.append(path);
    while len(queue) > 0:
        tmp = queue.pop() # push out the current item and use FIFO      
        if(os.path.isdir(tmp)):
            FileList = os.listdir(tmp)
            for item in FileList:
                queue.append(os.path.join(tmp, item))
        elif(os.path.isfile(tmp)):
            if str(tmp).endswith(".dll"):
                print(tmp)
   
    
try:
    BFS_Dir(r'C:\\')
except WindowsError:
        pass
```

5\. Uniform cost search

Consider a state space where the start state is 2 and each state k has three successors: numbers 2k, 2k+1, 2k+2. The cost from state k to each respective child is k, ground(k/2), k+2.

can you implement a uniform-cost-search algorithm with python. The goal state is number 85.

```python
import math
goal=85
state=[-1]*(goal+1)
state[2]=0
for k in range(2,int(goal/2+1)):
    if state[k]<0:continue# alreadly reach goal state
    for position, cost in[(k*2,k),(k*2+1,math.floor(k/2)),(k*2+2,k+2)]:
        if position>goal:continue # alreadly reach goal state
        if state[position]==-1 or state[position]>state[k]+cost:
            state[position]=state[k]+cost
print(state[goal])
```
